# Imports

```python
# Notebook Instance Imports
import os
import sagemaker
from sagemaker.tensorflow import TensorFlow
```

# Identify Data Location

```python
# S3 directories
data_s3 = 's3://jigsaw-toxic-mjy/'
#output_s3 = 's3://jigsaw-toxic-mjy-output/'
# define inputs
#inputs = {'data':data_s3, 'output':output_s3}
inputs = {'data':data_s3}
print(inputs)
```

# Create Training Image

```python
# create estimator
estimator = TensorFlow(entry_point='jigsaw_train1_aws2.py',
                       train_instance_type='ml.p2.xlarge',
                       output_path="s3://jigsaw-toxic-mjy-output",
                       train_instance_count=1,
                       role=sagemaker.get_execution_role(), # Passes to the container the AWS role that you are using on this notebook
                       framework_version='1.11.0', # Uses TensorFlow 1.11
                       py_version='py3',
                       script_mode=True)
```

# Run Training Job

```python
# Run training job
estimator.fit(inputs)
```
